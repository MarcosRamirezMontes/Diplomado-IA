```python
experiencia = [1,3,5,7,9,13]
salario = [8000,8500,9000,9500,10000,10500]
salario2 = [7900,8200,8800,9300,9800,10300]
```

```python
plt.plot(experiencia,salario, color = 'r')
plt.plot(experiencia,salario2, color = 'b')
plt.title('Salarios')
plt.xlabel('Años de experiencia')
plt.ylabel('Ganancias')
plt.legend(['salario','salario2'])
```

```python
plt.plot(experiencia,salario, color = 'r', linestyle = '--',marker = 'o')
plt.plot(experiencia,salario2, color = 'b')
plt.title('Salarios')
plt.xlabel('Años de experiencia')
plt.ylabel('Ganancias')
plt.legend(['salario','salario2'])
plt.grid(True, color = 'g')
```

```python
datos = [np.random.normal(0,std,100) for std in range (1,4)]
plt.boxplot(datos,vert=True)
```

```python
import pandas as pd
data_range = pd.date_range(start='2023-01-01',end='2023-12-31',freq='D')
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pd.DataFrame({'Date': data_range, 'Precios':precios})
data.head()
```

```python
plt.plot(data['Date'],data['Precios'], label='Precio')
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)

categorias = ['a','b','c','d']
valores= [5,7,3,8]
```

```python
fig, axs =plt.subplots(2,2, figsize = (10,10))
axs[0,0].plot(x,y1,color='r')
axs[0,0].set_title('Grafico de lineas')
axs[0,0].legend(['X'])
axs[0,0].grid()
'2.Graficos.ipynb'
axs[0,1].bar(categorias,valores,color='b')
axs[0,1].set_title('Grafico de barras')
axs[0,1].legend(['Valores'])


axs[1,0].plot(experiencia,salario,color='g')
axs[1,0].set_title('Histograma Salarios')
axs[1,0].legend(['Salario'])
axs[1,0].grid()


axs[1,1].scatter(xd,yd,color='r')
axs[1,1].set_title('Grafico de dispersion')
axs[1,1].legend(['yd'])

fig.suptitle('Ejemplo de diferentes gráficos',fontsize=20)
```
