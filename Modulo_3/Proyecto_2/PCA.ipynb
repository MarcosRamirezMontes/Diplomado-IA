# Ejemplo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
datos = pd.read_csv('advertising.csv')
datos.head()
```

#elegimos 2 componentes porque más abajo la varianza tenía un mayor valor en explained_variance = pd.DataFrame(pca.explained_variance_ratio) antes de especificar en la siguiente celda en PCA()

```python
modelo = Pipeline([
    ('scaler',StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('regresion',LinearRegression())
])
```

```python
X = datos[['TV','Radio','Newspaper']]
y = datos['Sales'] 
```

```python
modelo.fit(X,y)
```

```python
pca = modelo.named_steps['pca']
pca.components_
```

```python
explained_variance = pd.DataFrame(pca.explained_variance_ratio_)
explained_variance
```

```python
lr = modelo.named_steps['regresion']
lr.coef_
```

```python
lr.intercept_
```

```python
sales_pred = modelo.predict(X)
plt.scatter(datos['Sales'],sales_pred)
plt.plot([sales_pred.min(), sales_pred.max()], [sales_pred.min(), sales_pred.max()], '--r' )
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(datos['Sales'],sales_pred)
r2
```
